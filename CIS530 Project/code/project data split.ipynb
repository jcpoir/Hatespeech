```python
missing_df = joined_df[joined_df['implicit_class'].isna()]
missing_df = missing_df[missing_df['class_x']=='implicit_hate']
```

```python
missing_df = missing_df.drop(columns='extra_implicit_class').rename(columns={'class_x':'class'})
```

```python
implicit_other_df = joined_df[joined_df['implicit_class']=='other'].drop(columns='extra_implicit_class').rename(columns={'class_x':'class'})
```

```python
pre_split_df = joined_df[joined_df['implicit_class']!='other'].reset_index(drop=True)
```

```python
pre_split_df = pre_split_df.rename(columns={'class_x':'class'})
```

```python
pre_split_df['implicit_class'].value_counts()
```

```python
pre_split_df = pre_split_df.drop(columns='extra_implicit_class')
```

```python
pre_split_df
```

```python
set(pre_split_df['class'])
```

```python
def split(df):
  df_shuffled = df.sample(frac=1,random_state=25)
  train_df = df_shuffled[:len(df_shuffled)*8//10]
  dev_df = df_shuffled[len(df_shuffled)*8//10:len(df_shuffled)*9//10]
  test_df = df_shuffled[len(df_shuffled)*9//10:]
  return train_df,dev_df,test_df
```

```python
class_dfs = {}
for i in set(pre_split_df.dropna(subset=['implicit_class'])['implicit_class'].values):
  class_df  = pre_split_df[pre_split_df['implicit_class']==i]
  train_df,dev_df,test_df = split(class_df)
  class_dfs[i] = {
      'train' : train_df,
      'dev' : dev_df,
      'test' : test_df
  }
for j in ['explicit_hate','not_hate']:
  class_df = pre_split_df[pre_split_df['class'] == j]
  train_df,dev_df,test_df = split(class_df)
  class_dfs[j] = {
      'train' : train_df,
      'dev' : dev_df,
      'test' : test_df
  }

train_df_final = None
dev_df_final = None
test_df_final = None
empty_df = True
for post_class in class_dfs.keys():
  if empty_df:
    train_df_final = class_dfs[post_class]['train']
    dev_df_final = class_dfs[post_class]['dev']
    test_df_final = class_dfs[post_class]['test']
    empty_df = False
  else:
    train_df_final = pd.concat([train_df_final, class_dfs[post_class]['train']])
    dev_df_final = pd.concat([dev_df_final, class_dfs[post_class]['dev']])
    test_df_final = pd.concat([test_df_final, class_dfs[post_class]['test']])
```
